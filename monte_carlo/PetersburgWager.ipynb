```python
import matplotlib.pyplot as plt
import numpy as np
import math
import utilities
```

## The Cartoon Petersburg Wager

```python
payoffs = [1, 2, 6, 22, 200, 1000000]


wealths = np.linspace(1000, 150000, 100)

def fair_value(w):
    return utilities.bisection_root(lambda x: utilities.bev(x, w, payoffs)-1.0, 0.1, 1.0)

fair_values = np.asarray([fair_value(w) for w in wealths])

plt.plot(fair_values, wealths)
#plt.xlim(0, 1.0)
plt.ylim(0, 150000)
plt.xlabel("FRACTION OF STARTING WEALTH SHOULD WAGERED")
plt.ylabel("STARTING WEALTH")
plt.show()

starting_wealth = 100000
ymax = 160000
ratios = np.linspace(0.0, 1.0, 100)
ending_wealth = np.asarray([utilities.bev(x, starting_wealth, payoffs) * starting_wealth for x in ratios])
fv = fair_value(starting_wealth)

plt.plot(ratios, ending_wealth)
plt.xlim(0, 1.0)
plt.ylim(0, ymax)
plt.xlabel("FRACTION OF STARTING WEALTH WAGERED")
plt.ylabel("ENDING WEALTH")
plt.scatter(fv, starting_wealth, color='red')
plt.axvline(fv, linestyle='--', ymax=starting_wealth/ymax, color='gray')
plt.axhline(starting_wealth, xmax=fv, linestyle='--', color='gray')
plt.axhline()
plt.show()
```

##  Petersburg Wager

Initial wealth: $w$   
Costs: $c$  
Profit: $p$   
Loss rate: $l$    
Insurance cost: $i$
   

```python
w = 11000
c = 8000
p = 2000
l = 0.05
i = 800

geo_avg_no_ins = (1.0+p/w)**(1-l) * (1.0-c/w)**l
geo_avg_with_ins = (1.0+(p-i)/w)**(1-l) * (1.0-i/w)**l
geo_avg_with_ins_all = (1.0+(p-i)/w)

insurers_exp_profit = i * (1-l) - (p+w-i) * (l)

print(geo_avg_no_ins, geo_avg_with_ins, geo_avg_with_ins_all, insurers_exp_profit)
```
